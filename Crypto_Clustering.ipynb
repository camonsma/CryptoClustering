```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data) #similar to class exercise, scale and fit the numberic values
```

```python
# Create a DataFrame with the scaled data
df_market_transformed = pd.DataFrame(market_data_scaled, columns=["price_change_percentage_24h",\
                                                                   "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d",\
                                                                   "price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"])
#df_market_transformed.head()
# Copy the crypto names from the original data
coin_id_list = df_market_data.index.to_list()
#print(coin_id_list)
df_market_transformed["coin_id"] = coin_id_list
# Set the coinid column as index
df_market_transformed.set_index("coin_id",inplace=True)
# Display sample data
df_market_transformed.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
# exactly like class exercise for K means values for elbow fiting
# Create a a list to store inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Fit the model using the spread_df DataFrame
# Append the value of the computed inertia from the `inertia_` attribute of the KMeans model instance

df_market_data_scaled = df_market_transformed.copy()
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)

```

```python
# Create a dictionary with the data to plot the Elbow curve
# Exactly like class exercise
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame using the elbow_data Dictionary
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Cyrpto Currency Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4. Given the above visual I would elect to use a clustering of 4 because that appears to be the point where the line starts slow down on the decline and the y values appear to be bound to the value of 45.

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_lower = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(k_lower)
```

```python
# Create a copy of the DataFrame
spread_df_predictions = df_market_data_scaled.copy()
spread_df_predictions.head()
```

```python
# Add a new column to the DataFrame with the predicted clusters
#exactly like the class exercises.
spread_df_predictions['clusters_lower'] = k_lower

# Display sample data
#using the class exercise as a reference
spread_df_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
spread_df_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="clusters_lower",
    hover_cols = ['coin_id']
).opts(yformatter="%.0f")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
crypto_pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
#copy the original set into a new dataframe for more exploration
df_cyprto = df_market_transformed.copy()
ls_cyprto_pca = crypto_pca.fit_transform(df_cyprto)
# View the first five rows of the DataFrame. 
crypto_pca_df = pd.DataFrame(ls_cyprto_pca,columns=["PCA1","PCA2","PCA3"])
crypto_pca_df.head(5)
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
print('The explained variance')
crypto_pca.explained_variance_
print('The explained variance ratio')
crypto_pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three princinpal components is .372 + .347 + .176 = .895 or the model accounts for 89.5% of the oringinal data set.
